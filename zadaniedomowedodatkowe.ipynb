## Ćwiczenie 3 
- wyszukajcie posty na HN dot. interesującego tematu, korzystając z komórki powyżej
- zapisz metadane tych postów (tytuły, autorzy, punkty, linki, domeny) do list, następnie utwórzcie z nich data frame i csv
- przechodzenie na kolejne strony, url powinien wyglądać tak (tu dla przykładu "privacy"):

`https://hn.algolia.com/?dateRange=all&page=1&prefix=false&query=privacy&sort=byPopularity&type=story`
- iterujcie po `page=n`

```python
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import numpy as np  # Ensure you import numpy 
import pandas as pd
site = "https://hn.algolia.com/"

options = Options()
options.add_experimental_option("detach", True)
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)
driver.get(site)

titles = []
story_links = []
story_scores = []
domeny = []
autory = []

    
for n in range(6): # pobra 6 strony
    driver.get(site +'?dateRange=all&page='+str(n+1)+'&prefix=false&query=privacy&sort=byPopularity&type=story')
    tit = driver.find_elements(By.XPATH, '//div[1]/a[1]/span') #//div[1]/a[1]/span
    for t in tit:
        titles.append(t.text)
        
    link_elements = driver.find_elements(By.XPATH, '//article//div[1]/a[1]')
    for element in link_elements:
        story_links.append(element.get_attribute('href'))
        
        
    dom=driver.find_elements(By.XPATH, "//div[1]/a[2][@href]")
    for d in dom:
        try:
            domen=d.get_attribute('href')
            domeny.append(domen)
        except:
            domeny.append(np.nan)
    
    st=driver.find_elements(By.XPATH, "//div[2]/span[1]/a[1]")
    for s in st:
        try:
            score=s.text.split()[0]
            story_scores.append(score)
        except:
            story_scores.append(np.nan)

    aut=driver.find_elements(By.XPATH, " //span[3]/a/span")
    for a in aut:
        try:
            autor=a.text
            autory.append(autor)
        except:
            autory.append(np.nan)
    sleep(1)
```
